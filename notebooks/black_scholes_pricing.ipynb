```python
import numpy as np
import matplotlib.pyplot as plt
from utils.bs_functions import black_scholes_price, delta, gamma, vega

# Single Option Pricing


S = 100      # current stock price
K = 100      # strike price
r = 0.05     # risk-free rate
T = 1.0      # time to maturity (1 year)
sigma = 0.2  # volatility

call_price = black_scholes_price(S, K, r, T, sigma, 'call')
put_price = black_scholes_price(S, K, r, T, sigma, 'put')

print(f"Call Price: {call_price:.2f}")
print(f"Put Price: {put_price:.2f}")

# Plotting Call Price vs Strike
K_vals = np.linspace(50, 150, 100)
call_prices = [black_scholes_price(S, K, r, T, sigma, 'call') for K in K_vals]

plt.figure(figsize=(10, 5))
plt.plot(K_vals, call_prices, label='Call Option Price')
plt.xlabel('Strike Price (K)')
plt.ylabel('Option Price')
plt.title('Call Price vs Strike (Black-Scholes)')
plt.grid(True)
plt.legend()
plt.show()


# Delta + Vega Visualization

S_vals = np.linspace(60, 140, 100)
deltas = [delta(S, K, r, T, sigma, 'call') for S in S_vals]
vegas = [vega(S, K, r, T, sigma) for S in S_vals]

plt.figure(figsize=(10, 5))
plt.plot(S_vals, deltas, label='Delta')
plt.plot(S_vals, vegas, label='Vega')
plt.xlabel('Stock Price (S)')
plt.ylabel('Sensitivity')
plt.title('Option Sensitivities (Greeks)')
plt.grid(True)
plt.legend()
plt.show()
```
